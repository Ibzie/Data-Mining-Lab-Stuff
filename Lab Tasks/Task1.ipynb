```python
Movies_Data.dropna(inplace=True)
Movies_Data.isna()
```

```python
Fifa_Dataset.isna().sum().sum()
```

```python
numerical_columns = Fifa_Dataset.select_dtypes(include=['number']).columns
Fifa_Dataset[numerical_columns] = Fifa_Dataset[numerical_columns].apply(lambda x: x.fillna(x.mean()))
```

```python
# Replace NaN values in text columns with the median string
text_columns = Fifa_Dataset.select_dtypes(include=['object']).columns
Fifa_Dataset[text_columns] = Fifa_Dataset[text_columns].apply(lambda x: x.fillna(x.mode().iloc[0]))
```

```python
Fifa_Dataset.isna().sum().sum()
```

```python
Movies_Data['Votes'] = pd.to_numeric(Movies_Data['Votes'], errors='coerce')
top_10_movies = Movies_Data.nlargest(10, ['Rating', 'Votes'])

print(top_10_movies)
```

```python
Movies_Data['Profit'] = Movies_Data['Revenue (Millions)'] - Movies_Data['Revenue (Millions)']
top_10_movies = Movies_Data.sort_values(by='Profit', ascending=False).head(10)
top_10_movies
```

```python
#Is it related? idk lets see

import matplotlib.pyplot as plt

plt.scatter(Movies_Data['Rating'], Movies_Data['Revenue (Millions)'])
plt.title('Movie Rating vs Revenue')
plt.xlabel('Rating')
plt.ylabel('Revenue (Millions)')
plt.show()

correlation_coefficient = Movies_Data['Rating'].corr(Movies_Data['Revenue (Millions)'])

print(f'Correlation Coefficient: {correlation_coefficient}')
```

I guess it is related sometimes but not indefinitely

```python
director_name = 'Bryan Singer'
movies_by_director = Movies_Data[Movies_Data['Director'] == director_name]
average_runtime = movies_by_director['Runtime (Minutes)'].mean()

print(f'The average runtime of movies directed by {director_name} is: {average_runtime:.2f} minutes')
```

```python
Movies_Data['Genre'] = Movies_Data['Genre'].str.split(',')
exploded_df = Movies_Data.explode('Genre') #Explode
average_rating_by_genre = exploded_df.groupby('Genre')['Rating'].mean()

highest_rated_genre = average_rating_by_genre.idxmax()
highest_average_rating = average_rating_by_genre.max()

print(f'The genre with the highest average rating is {highest_rated_genre} with an average rating of {highest_average_rating:.2f}')
```

```python
target_year = 2010
movies_in_year = Movies_Data[Movies_Data['Year'] == target_year]
total_revenue = movies_in_year['Revenue (Millions)'].sum()

print(f'The total revenue generated by movies released in {target_year} is: ${total_revenue:.2f} million')
```

```python
director_count = Movies_Data['Director'].value_counts()
most_movies_directed = director_count.idxmax()
num_movies_directed = director_count.max()

print(f'The director is {most_movies_directed} with {num_movies_directed} movies.')
```

```python
plt.scatter(Movies_Data['Runtime (Minutes)'], Movies_Data['Rating'])
plt.title('Runtime vs Rating')
plt.xlabel('Runtime (Minutes)')
plt.ylabel('Rating')
plt.show()


correlation_coefficient = Movies_Data['Runtime (Minutes)'].corr(Movies_Data['Rating'])
print(f'Correlation Coefficient between Runtime and Rating: {correlation_coefficient}')
```

I couldnt see the pattern based on correlation but the coefficient suggests its not too related
